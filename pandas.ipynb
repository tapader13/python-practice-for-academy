```python
#dataframe
import pandas as pd

data={
    "Name":["minhaj","miftah","mizan","marzan"],
    "Age":[23,20,17,12],
    "Gander":["male","male","male","male"]
}

df=pd.DataFrame(data)
print(df)
print(df.loc[0])

df1=pd.DataFrame(data,index=["brother1","brother2","brother3","brother4"])

#loc
print(df1)
print(df1.loc[["brother1"]])
print(df1.loc[["brother1","brother2"]])

print(df1.loc[:,["Name"]])
print(df1.loc[:,["Name","Age"]])
print(df1.loc[["brother1","brother2"],["Name","Age"]])

#iloc
print(df1.iloc[[0]])
print(df1.iloc[[0,1]])
print(df1.iloc[:,[0]])
print(df1.iloc[:,[0,1]])
print(df1.iloc[[0,1],[0,1]])

#iterate
for i in df:
  print(i)
for i in df:
  print(df.loc[:,[i]])


#attributes & methods
print(df1.dtypes)
print(df1.shape)
print(df1.size)
print(df1.head())
print(df1.head(2))
print(df1.tail())
print(df1.tail(2))
print(df1.ndim)
print(df1.index)
print(df1.T)
```

```python
#join
import pandas as pd

# First DataFrame
data1 = {
    "Name": ["Minhaj", "Miftah", "Mizan", "Marzan","test"],
    "Age": [23, 20, 17, 12,23]
}
df1 = pd.DataFrame(data1)

# Second DataFrame
data2 = {
    "City": ["Dhaka", "Chittagong", "Khulna", "Rajshahi","test1","test2"],
    "Country": ["Bangladesh", "Bangladesh", "Bangladesh", "Bangladesh","ind","pak"]
}
df2 = pd.DataFrame(data2)

# Simple join (both have same default index 0,1,2,3)
df3 = df1.join(df2)

print(df3)

df4 = df1.join(df2,how="left")
print(df4)

df5 = df1.join(df2,how="right")
print(df5)

df6 = df1.join(df2,how="inner")
print(df6)

df7 = df1.join(df2,how="outer")
print(df7)


#concatinate
data3 = {
    "Name":["minhaj","miftah","mizan","marzan"],
    "Age":[23,20,17,12],
    "Gander":["male","male","male","male"]
}
data4 = {
    "Name":["tanbir","zakir"],
    "Age":[23,45],
    "Gander":["male","male"]
}
df8=pd.DataFrame(data3)
df9=pd.DataFrame(data4)
df10=pd.concat([df8,df9])
print(df10)

```

```python
#series
import pandas as pd
import numpy as np

data=[1,2,3,4,5,None]
s=pd.Series(data,index=["s1","s2","s3","s4","s5","s6"],name="test series")
print(s)

#access
print(s[0])
print(s["s2"])
print(s.loc["s2"])
print(s.iloc[1])
print(s.loc[["s2","s3"]])
print(s.iloc[[1,2]])
print(s["s1":"s4"])
print(s[0:4])
print(s.loc["s1":"s4"])
print(s.iloc[0:4])


#attribute & methods
print(s.shape)
print(s.size)
print(s.ndim)
print(s.head())
print(s.head(2))
print(s.tail())
print(s.tail(2))
print(s.index)
print(s.values)
print(s.dtypes)
print(s.name)
print(s.hasnans)
print(s.info())

#combine
s1=pd.Series([1,15,3,4],index=["s1","s2","s3","s4"])
s2=pd.Series([11,12,13,14,15],index=["s1","s2","s3","s4","s5"])

s3=s1.combine(s2,lambda x,y:x+y)
print(s3)

s4=s1.combine(s2,lambda x,y:x*y)
print(s4)

s5=s1.combine(s2,np.divide)
print(s5)

def findMax(x1,x2):
  if(x1>x2):
    return x1
  else:
    return x2
s6=s1.combine(s2,findMax,fill_value=20) #fill value will replace none value to fill it
print(s6)
```
