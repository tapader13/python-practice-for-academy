```python
#dataframe
import pandas as pd

data={
    "Name":["minhaj","miftah","mizan","marzan"],
    "Age":[23,20,17,12],
    "Gander":["male","male","male","male"]
}

df=pd.DataFrame(data)
print(df)
print(df.loc[0])

df1=pd.DataFrame(data,index=["brother1","brother2","brother3","brother4"])

#loc
print(df1)
print(df1.loc[["brother1"]])
print(df1.loc[["brother1","brother2"]])

print(df1.loc[:,["Name"]])
print(df1.loc[:,["Name","Age"]])
print(df1.loc[["brother1","brother2"],["Name","Age"]])

#iloc
print(df1.iloc[[0]])
print(df1.iloc[[0,1]])
print(df1.iloc[:,[0]])
print(df1.iloc[:,[0,1]])
print(df1.iloc[[0,1],[0,1]])

#iterate
for i in df:
  print(i)
for i in df:
  print(df.loc[:,[i]])


#attributes & methods
print(df1.dtypes)
print(df1.shape)
print(df1.size)
print(df1.head())
print(df1.head(2))
print(df1.tail())
print(df1.tail(2))
print(df1.ndim)
print(df1.index)
print(df1.T)
```

```python
#join
import pandas as pd

# First DataFrame
data1 = {
    "Name": ["Minhaj", "Miftah", "Mizan", "Marzan","test"],
    "Age": [23, 20, 17, 12,23]
}
df1 = pd.DataFrame(data1)

# Second DataFrame
data2 = {
    "City": ["Dhaka", "Chittagong", "Khulna", "Rajshahi","test1","test2"],
    "Country": ["Bangladesh", "Bangladesh", "Bangladesh", "Bangladesh","ind","pak"]
}
df2 = pd.DataFrame(data2)

# Simple join (both have same default index 0,1,2,3)
df3 = df1.join(df2)

print(df3)

df4 = df1.join(df2,how="left")
print(df4)

df5 = df1.join(df2,how="right")
print(df5)

df6 = df1.join(df2,how="inner")
print(df6)

df7 = df1.join(df2,how="outer")
print(df7)


#concatinate
data3 = {
    "Name":["minhaj","miftah","mizan","marzan"],
    "Age":[23,20,17,12],
    "Gander":["male","male","male","male"]
}
data4 = {
    "Name":["tanbir","zakir"],
    "Age":[23,45],
    "Gander":["male","male"]
}
df8=pd.DataFrame(data3)
df9=pd.DataFrame(data4)
df10=pd.concat([df8,df9])
print(df10)

```

```python
#sort
import numpy as np
import pandas as pd
data={
    "Name":["minhaj","miftah","mizan","marzan"],
    "Age":[23,20,17,12],
    "Gander":["male","male","male","male"]
}
df=pd.DataFrame(data)
print(df)
print(df.sort_values(by=["Name","Age"]))
print(df.sort_values(by=["Age"], ascending=False))

#duplicate
data1={
    "Name":["minhaj","miftah","mizan","marzan","minhaj"],
    "Age":[23,20,17,12,23],
    "Gander":["male","male","male","male","male"]
}
df1=pd.DataFrame(data1)
print(df1)
print(df1.duplicated())
print(df1.drop_duplicates())

#clean data
data2={
    "Name":["minhaj","miftah","mizan","marzan",np.nan],
    "Age":[23,20,17,12,23],
    "Gander":["male","male","male","male","male"]
}
df2=pd.DataFrame(data2)
print(df2)
print(df2.isnull())
print(df2.notnull())
print(df2.dropna())
print(df2.fillna(5))
res=df2.fillna(4)
print(res.to_string())
```

```python
# groupby
import pandas as pd

data={
    "Name":["minhaj","miftah","mizan","marzan","mona","kona"],
    "Age":[23,20,17,12,67,76],
    "Gander":["male","male","male","male","female","female"]
}
df=pd.DataFrame(data)
groupRes=df.groupby("Gander")
print(groupRes)
print(groupRes.groups)
for gName,gValue in groupRes:
  print(gName)
  print(gValue)

print(groupRes["Age"].sum())
print(groupRes["Age"].mean())
print(groupRes["Age"].size())
print(groupRes["Age"].agg(max))
print(groupRes["Age"].agg([max,min,"mean","count"]))
```
