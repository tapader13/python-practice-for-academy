```python
#basic plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x= np.array([1,2,3,4,5])
y= np.array([1,4,9,16,25])
plt.plot(x,y,color="red",marker="o",label="random plot chart",linestyle=":",linewidth=1) #for lebel must call legend
plt.legend(loc="upper right",fontsize=10)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.xlim(0,6)
plt.ylim(0,30)
plt.xticks([1,2,3,4,5],["num1","num2","num3","num4","num5"])
plt.yticks([1,4,9,16,25],["sqr1","sqr2","sqr3","sqr4","sqr5"])
plt.savefig("lineplot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#bar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
products=np.array(["alu","begun","piyaj"])
sales=np.array([20,40,34])
plt.bar(products,sales,color="green",width=.5,label="product vs sales comparesion")
plt.title("bar chart")
plt.legend(loc="upper right",fontsize=10)
plt.xlabel("products")
plt.ylabel("sales")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.savefig("barplot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#horizontal bar
#bar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
products=np.array(["alu","begun","piyaj"])
sales=np.array([20,40,34])
plt.barh(products,sales,color="green",height=.5,label="product vs sales comparesion")
plt.title("bar chart")
plt.legend(loc="upper right",fontsize=10)
plt.xlabel("products")
plt.ylabel("sales")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.savefig("barplot_horizontal.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#pie
import numpy as np
import matplotlib.pyplot as plt
labels=np.array(["minhaj","miftah","mizan","marzan"])
values=np.array([50,54,76,78])
plt.pie(values,labels=labels,autopct="%.1f%%")
plt.legend(loc="upper right",fontsize=10)
plt.title("pie chart")
plt.savefig("pieplot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#histogram
import numpy as np
import matplotlib.pyplot as plt
values=np.random.randint(1,100,50)
plt.hist(values,bins=7,color="green",edgecolor="red",density=True)
plt.title("histogram")
plt.xlabel("values")
plt.ylabel("frequency")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.savefig("histogram.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#scatter
import numpy as np
import matplotlib.pyplot as plt
study=np.random.randint(5,20,10)
result=np.random.randint(2,20,10)
plt.scatter(study,result,color="blue",marker="o",label="study vs result")
plt.title("scatter plot")
plt.legend(loc="upper right",fontsize=10)
plt.xlabel("study")
plt.ylabel("result")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.savefig("scatterplot.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#subplot
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.subplot(1,2,1)
plt.plot(x,y,color="red",marker="o",label="random plot chart",linestyle=":",linewidth=1)
plt.legend(loc="upper right",fontsize=10)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(1,2,2)
plt.bar(x,y,color="red",label="random plot chart",linestyle=":",linewidth=1)
plt.legend(loc="upper right",fontsize=10)
plt.title("Bar Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.tight_layout() #automatically adjust spacing between 2 subplot
plt.savefig("subplot.png",dpi=300,bbox_inches="tight")

plt.show()
```

```python
#subplot using oo api
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])

fig,axes=plt.subplots(1,2,figsize=(10,4))
fig.suptitle("subplot")
axes[0].plot(x,y,color="red",marker="o",label="random plot chart",linestyle=":",linewidth=1)
axes[0].legend(loc="upper right",fontsize=10)
axes[0].set_title("Line Plot")
axes[0].set_xlabel("x-axis")
axes[0].set_ylabel("y-axis")


axes[1].bar(x,y,color="red",label="random plot chart",linestyle=":",linewidth=1)
axes[1].legend(loc="upper right",fontsize=10)
axes[1].set_title("Bar Plot")
axes[1].set_xlabel("x-axis")
axes[1].set_ylabel("y-axis")

fig.tight_layout() #automatically adjust spacing between 2 subplot
fig.savefig("subplot.png",dpi=300,bbox_inches="tight")

plt.show()
```

```python
#netflix data
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/drive/MyDrive/netflix1.csv")
print(df.head())
# print(df.columns)
# print(df.info())
# print(df.describe())
# print(df.isnull().sum())
df=df.dropna(subset=["show_id","date_added","release_year","rating","duration"])
type_counts=df["type"].value_counts()
print(type_counts)

plt.pie(type_counts,labels=type_counts.index,autopct="%.1f%%")
plt.title("pie chart")
plt.savefig("netflixtypeshow.png",dpi=300,bbox_inches="tight")
plt.show()

country_counts=df["country"].value_counts()
print(country_counts)

plt.figure(figsize=(14,5))
plt.bar(country_counts.index,country_counts.values,color="green",width=.5,label="country counts comparesion")
plt.title("bar chart")
plt.legend(loc="upper right",fontsize=10)
plt.xlabel("country")
plt.ylabel("counts")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("netflixcountry.png",dpi=300,bbox_inches="tight")
plt.show()
```
