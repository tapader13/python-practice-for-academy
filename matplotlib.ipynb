```python
#basic plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x= np.array([1,2,3,4,5])
y= np.array([1,4,9,16,25])
plt.plot(x,y,color="red",marker="o",label="random plot chart",linestyle=":",linewidth=1) #for lebel must call legend
plt.legend(loc="upper right",fontsize=10)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True,color="green",linestyle="-.",linewidth=.5)
plt.xlim(0,6)
plt.ylim(0,30)
plt.xticks([1,2,3,4,5],["num1","num2","num3","num4","num5"])
plt.yticks([1,4,9,16,25],["sqr1","sqr2","sqr3","sqr4","sqr5"])
plt.savefig("lineplot.png",dpi=300,bbox_inches="tight")
plt.show()
```
